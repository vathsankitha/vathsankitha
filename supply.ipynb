```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 📅 Create synthetic dataset
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

np.random.seed(42)  # reproducibility
demand = np.random.randint(800, 1500, size=12)
supply = demand + np.random.randint(-300, 300, size=12)

df = pd.DataFrame({
    'Month': months,
    'Demand': demand,
    'Supply': supply
})
df['Gap'] = df['Demand'] - df['Supply']

# 📈 Analysis
max_demand_month = df.loc[df['Demand'].idxmax(), 'Month']
min_demand_month = df.loc[df['Demand'].idxmin(), 'Month']
max_supply_month = df.loc[df['Supply'].idxmax(), 'Month']
min_supply_month = df.loc[df['Supply'].idxmin(), 'Month']

print(f"📌 Max Demand: {df['Demand'].max()} in {max_demand_month}")
print(f"📌 Min Demand: {df['Demand'].min()} in {min_demand_month}")
print(f"📌 Max Supply: {df['Supply'].max()} in {max_supply_month}")
print(f"📌 Min Supply: {df['Supply'].min()} in {min_supply_month}")

# 🎨 Set style
sns.set(style="whitegrid")
plt.figure(figsize=(14, 12)) # Increased figure size to accommodate more plots

# 1️⃣ Line Chart
plt.subplot(4, 2, 1) # Changed subplot grid
plt.plot(df['Month'], df['Demand'], label='Demand', marker='o')
plt.plot(df['Month'], df['Supply'], label='Supply', marker='s')
plt.title('Line Chart')
plt.legend()

# 2️⃣ Bar Chart
plt.subplot(4, 2, 2) # Changed subplot grid
x = np.arange(len(df['Month']))
plt.bar(x - 0.2, df['Demand'], width=0.4, label='Demand')
plt.bar(x + 0.2, df['Supply'], width=0.4, label='Supply')
plt.xticks(x, df['Month'])
plt.title('Bar Chart')
plt.legend()

# 3️⃣ Pie Chart (Demand Share)
plt.subplot(4, 2, 3) # Changed subplot grid
plt.pie(df['Demand'], labels=df['Month'], autopct='%1.1f%%')
plt.title('Demand Distribution (Pie)')

# 4️⃣ Scatter Plot
plt.subplot(4, 2, 4) # Changed subplot grid
plt.scatter(df['Demand'], df['Supply'], c='purple')
plt.xlabel('Demand')
plt.ylabel('Supply')
plt.title('Scatter Plot')

# 5️⃣ Area Chart
plt.subplot(4, 2, 5) # Changed subplot grid
plt.stackplot(df['Month'], df['Demand'], df['Supply'], labels=['Demand', 'Supply'], alpha=0.6)
plt.title('Area Chart')
plt.legend()

# 6️⃣ Histogram
plt.subplot(4, 2, 6) # Changed subplot grid
plt.hist(df['Gap'], bins=6, color='orange', edgecolor='black')
plt.title('Demand-Supply Gap Histogram')

# 7️⃣ Boxplot
plt.subplot(4, 2, 7) # Changed subplot grid
sns.boxplot(data=df[['Demand', 'Supply']])
plt.title('Boxplot of Demand & Supply')

# 8️⃣ Stacked Bar Chart
plt.subplot(4, 2, 8) # Added new subplot
plt.bar(df['Month'], df['Demand'], label='Demand')
plt.bar(df['Month'], df['Supply'], bottom=df['Demand'], label='Supply')
plt.title('Stacked Bar Chart')
plt.legend()

# # 9️⃣ Violin Plot (Optional - can add if needed)
plt.subplot(4, 2, 8) # Changed subplot grid - adjust if adding violin plot
sns.violinplot(data=df[['Demand', 'Supply']])
plt.title('Violin Plot of Demand & Supply')


# 🖼️ Layout and show
plt.tight_layout()
plt.show()
```
